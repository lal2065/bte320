in class assignment 4 - for loop

```python
fixed_cost = 200
initial_attendees = 20
ticket_price = 10

print("Ads $ \t Profit")
for ad_dollars in range(0, 225, 25):
  additional_attendees = (2 * round(ad_dollars**(1/2)))
  profit = ticket_price * (additional_attendees + initial_attendees) - ad_dollars - fixed_cost
  print(f"{ad_dollars}\t{profit}")
```

in class assignment 5 - while loop

```python
print("Hour\tFee")
hours = 1
while hours < 9:
  fee = 5 + (2.5 * hours)
  if fee < 10:
    fee = 10
  elif fee > 20:
    fee = 20
  print(f"{float(hours)}\t{float(fee)}")
  hours += 1
```

in class assignment 6 - rock paper scissors

```python
import random
check = input("Do you want to play (y/n)?").lower().strip()
while check == 'y':
  player_move = input("Enter a choice (rock, paper, scissors): ").lower().strip()
  print("The computer chose", computer_move)
  print(f"Player: {player_move}\nComputer: {computer_move}")
  if player_move == computer_move:
    print("tie")
  elif (
      (player_move == 'rock' and computer_move == 'scissors') or
       (player_move == 'paper' and computer_move == 'rock') or
        (player_move == 'scissors' and computer_move == 'paper')
        ):
    print("you win")
  else:
    print("you lose")
  check = input("Do you want to continue (y/n)?").lower().strip()
```
