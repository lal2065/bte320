in class assignment 12: greeting class

```python
class Greeting:
  def __init__(self, first, last):
    self.first = first
    self.last = last
  def display(self, formal=False):
    if formal:
      print(f"{self.last}, {self.first}")
    else:
      print(f"{self.first}")
  def __str__(self):
    return "Instance of Class 'Greeting'"

name = Greeting("John", "Doe")
print(name)
name.display(True)
name.display()
```

in class assignment 13

```python
class Student:
  def __init__(self, name, number):
    self.__name = name
    self.__number = number
    self.__courses = []
  def get_name(self):
    return self.__name
  def set_name(self, newName):
    self.__name = newName
  def get_number(self):
    return self.__number
  def set_number(self, newNumber):
    self.__number = newNumber
  def get_courses(self):
    return self.__courses
  def __len__(self):
    return len(self.__courses)
  def enroll(self,newCourse):
    if newCourse not in self.__courses:
      self.__courses.append(newCourse)
      print(f"enrolled in {newCourse}")
    else:
      print(f"Student {self.__name} already enrolled in {newCourse}")
```

```python
s = Student("Jane Doe", 12345)
print(f"Student name:", s.get_name())
print(f"Student number:", s.get_number())
print(f"Student courses:", s.get_courses())

s.enroll("Calculus I")
s.enroll("Physics I")
s.enroll("Calculus I") # This should show that she's already enrolled

print(f"Number of courses enrolled: {len(s)}")
print(f"Updated student courses:", s.get_courses())
```

in class assignment 14 -  class point and class rectange

```python
class Point:
  def __init__(self,x,y):
    self.x = x
    self.y = y
  def distance(self, other):
    dx = self.x - other.x
    dy = self.y - other.y
    return ((dx**2 + dy**2)**0.5)

class Rectangle:
  def __init__(self, point1, point2, point3, point4):
    self.point1 = point1
    self.point2 = point2
    self.point3 = point3
    self.point4 = point4
  def width(self):
    return self.point1.distance(self.point2)
  def height(self):
    return self.point1.distance(self.point3)
  def area(self):
    return self.width() * self.height()
```

```python
class Employee:
  def __init__(self, name, hourly_wage, hours_worked):
    self.name = name
    self.__hourly_wage_ = hourly_wage
    self.hours_worked = hours_worked
  def calculate_pay(self):
    return self.__hourly_wage_ * self.hours_worked
  def __str__(self):
    return f"Name: {self.name}, Hourly Wage: {self.__hourly_wage_}, Hours Worked: {self.hours_worked}"
  def get_wage(self):
    return self.__hourly_wage_
  def set_wage(self,new_wage):
    assert new_wage > 0, "Negative wages not allowed"
    self.__hourly_wage_ = new_wage

class Manager(Employee):
  def __init__(self, name, hourly_wage, hours_worked, bonus):
    super().__init__(name, hourly_wage, hours_worked)
    self.bonus = bonus

  def calculate_pay(self):
    return (1 + self.bonus) * self.get_wage() * self.hours_worked


employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, 0.2)
print(employee)
print(manager)
print()
print(f"Employee {employee.name} weekly salary is {employee.calculate_pay()}$")
print(f"Manager {manager.name} weekly salary is {manager.calculate_pay()}$ that includes a {manager.bonus * 100}% bonus")
```
